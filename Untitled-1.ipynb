```python
import networkx as nx

from bokehgraph import BokehGraph
```

```python
g = nx.Graph()
g.add_node(1, bipartite=0, gender="m")
g.add_node(2, bipartite=0, gender="w")
g.add_node(3, bipartite=1, gender="m")
g.add_node(4, bipartite=1, gender="w")
g.add_edge(1, 3)
g.add_edge(1, 4)
g.add_edge(2, 3)
g.add_edge(2, 4)

plot = BokehGraph(
    g,
    width=500,
    height=500,
    hover_edges=True,
    hover_nodes=True,
    inline=True,
    bipartite=True,
)
fig = plot.draw(node_size=20, return_figure=True)
```
